```python
df['Release Date'] = pd.to_datetime(df['Release Date'])
df['Release Year'] = df['Release Date'].dt.year
df['Release Month'] = df['Release Date'].dt.month
df['Release Month Name'] = df['Release Date'].dt.month_name()
df['Release Day'] = df['Release Date'].dt.day
df['Release Day of Week'] = df['Release Date'].dt.dayofweek
df['Release Day Name'] = df['Release Date'].dt.day_name()

df.head()
```

# Data Visualization And Analysis

# New to town

```python
if 'Artist' in df.columns:
    top_artists = df['Artist'].value_counts().head(10)

    plt.figure(figsize=(12, 6))
    sns.barplot(x=top_artists.index, y=top_artists.values)
    plt.title('Top 10 Most Frequent Artists')
    plt.xlabel('Artist')
    plt.ylabel('Number of Songs')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Column 'Artist' not found in the dataset.")
```

```python

if 'Duration' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.histplot(df['Duration'], bins=50, kde=True)
    plt.title('Distribution of Song Durations')
    plt.xlabel('Duration (seconds)')
    plt.ylabel('Count')
    plt.show()
else:
    print("Column 'Duration' not found in the dataset.")
```

# Songs Popularity:

```python
if 'Popularity' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.histplot(df['Popularity'], bins=50, kde=True)
    plt.title('Distribution of Song Popularity')
    plt.xlabel('Popularity')
    plt.ylabel('Count')
    plt.show()
else:
    print("Column 'Popularity' not found in the dataset.")
```

```python
sns.boxplot(df, x='Popularity', y='Genre')
plt.title('Popularity by Genre');
```

# Popularity Achieved

```python
sns.set_theme(style="darkgrid")

plt.figure(figsize=(15,10))
sns.barplot(df.groupby('Artist', as_index=False)['Album'].count().sort_values(by='Album',ascending=False).head(50),y='Artist',x='Album')
plt.title('Top 50')
plt.show()
```
